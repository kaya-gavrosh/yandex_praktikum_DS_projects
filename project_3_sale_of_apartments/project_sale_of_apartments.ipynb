# Исследование объявлений о продаже квартир

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Цель исследования**

Цель исследования - провести исследовательский анализ данных для установления параметров, влияющих на стоимость объектов недвижимости.

Для этого предстоит изучить следующие вопросы:
- Изучить такие параметры как площадь, цена, число комнат, высота потолков.
- Проанализировать время, затраченное на продажу квартиры.
- Изучить факторы, влияющие на стоимость квартиры в большей степени (число комнат, этаж, площадь, удаленность от центра, дата размещения).
- Найти населенные пункты с самой высокой и низкой стоимостью жилья.
- Определить центральную зону города и проанализировать квартиры в этой области.

**Задачи**

В ходе исследования необходимо рассчитать следующие параметры:
- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры (первый, последний, другой);
- соотношение жилой и общей площади, а также отношение площади кухни к общей.



**Ход исследования**

Данные о квартирах содержатся в файле `real_estate_data.csv`. Поскольку качество данных неизвестно, необходимо провести обзор данных перед началом исследования.

 1. Обзор данных.

### Этаж квартиры

Добавим в таблицу данные о том, на каком этаже: первом, последнем или другом – находится квартира. При этом практически все квартиры, для которых не было указано число этажей в доме, попадут в группу "другой", поскольку при заполнении пропусков в этой колонке выяснилось, что только три квартры находятся на 1 этаже. К счастью, пропусков было немного, скорее всего, они не сильно повлияют на результат.

Напишем функцию, которая будет одно из трех значений,  в зависимости от того, на каком этаже квартира.

```python
# функця flat_floor
def flat_floor(row):
    
    ''' 
    возвращает 'первый', если квартира на 1 этаже; 
    'последний', если этаж квартиры равен числу этажей в доме и квартира не на первом этаже,
    'другой' для остальных случаев
    '''
    
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] != 1 and row['floors_total'] == row['floor']:
        return 'последний'
    else:
        return 'другой'
```

```python
# проверка функции 
flat_floor(df.loc[186, :])
```

```python
# поиск подходящих по условию строк для проверки функции
df.query('floor == floors_total').head(3)
```

Функция работает исправно при проверки на отдельных строках с заранее известными параметрами. Применим функцию ко всей таблице и запишем полученные результаты в новый столбец

```python
# применение функции flat_floor
df['flat_floor'] = df.apply(flat_floor, axis=1)
```
