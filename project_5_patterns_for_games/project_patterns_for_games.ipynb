### Типы данных

Рассмотрим столбец `user_score`.

Обработка кода в ячейке выше прерывается на значение 'tbd' ("подлежит определению") Посмотрим, сколько строк содержат его.

```python
df[df['user_score'] == 'tbd']['name'].count() / len(df) * 100
```

Это значение содержится в 14,5% строк. Сделаем срез по этим строкам для отдельного анализа.

```python
# срез данных
tbd = df[df['user_score'] == 'tbd'].copy()
```

Проверим, есть ли еще строковые значения в этом столбце

```python
# уникальные значения столбца user_score
df.query('user_score != "tbd"')['user_score'].unique()
```

Больше строковых объектов нет. Перейдем к рассмотрению строк со значением "tbd".

```python
# получение первых строк
tbd.head()
```

```python
# число строк в срезе tbd
len(tbd)
```

Бросается в глаза, что  продажи в Японии равны нулю, проверим так ли это.

```python
tbd.query('jp_sales != 0')['jp_sales'].count()
```

Все же есть значения, которые больше нуля. Посмотрим на другие столбцы. Быть может, значение "tbd" в столбце с пользовательскими оценками, как-то связано со значениями других признаков.

```python
# уникальные значения столбца platform
tbd['platform'].unique()
```

```python
# максимальный и минимальный год в столбце year_of_release среза tbd
print('Самый ранний год:', tbd['year_of_release'].min())
print('Самый поздний год:', tbd['year_of_release'].max())
```

```python
# чистло пропусков в столбце critic_score среза tbd
tbd.query('critic_score.isna()')['name'].count()
```

```python
# чистло пропусков в столбце rating среза tbd
tbd.query('rating.isna()')['name'].count()
```

```python
# уникальные значения столбца genre
tbd['genre'].unique()
```

Ничего примечательного в других столбцах не обнаружено.

Видимо, в строках, в которых вместо числового значения стоит "tbd", пользовательский рейтинг не определен. То есть это нельзя назвать пропуском, поскольку значение в принципе не было установлено. Эти строки надо рассматривать (в контексте пользоваетльских рейтингов) отдельно. Чтобы строковое значение не мешало при дальнейшей работе, заменим его на техническое значение "-1", которое будет показывать не отрицательный рейтинг (поскольку в данном случае это невозможно), а группу игр, для которой не установлено значение пользовательских оценок.

```python
# замена значений в столце user_score
df['user_score'] = df['user_score'].replace('tbd', -1)
```

Вернемся к замене типа данных в этом столбце

```python
# изменение строкового типа на числовой
df['user_score'] = pd.to_numeric(df['user_score'], errors='raise')
```

Тип данных в столбце `year_of_release` заменим после обработки пропусков.

**Вывод**

Значения столбца `user_score` были заменены на тип `float`. Там же было обнаружено строковое значение "tbd" , которое было заменено на "-1". Эти строки обозначают то, что пользовательский рейтинг для игр в целом не был определен.

#### year_of_release

Пропусков в этой колонке не так много (менее 2%). Посмотрим на строки с пропусками в этом столбце.

```python
# получение первых строк
df.query('year_of_release.isna()').head()
```

В некоторых названиях игр указан год выхода (как правило, игра выходит в конце предыдущего года), т.е. для некоторых строк можно будет восстановить данные. 

Сделаем функцию, которая будет вычленять год выпуска игры из названия.

```python
# функция search_of_year
def search_of_year(row):
    """
    функция разбивает название игры по пробелам и, если последний элемент является
    числом, большим/равным 1980 (чтобы исключить исторические игры, части и т.д.), 
    возвращает предыдущий год
    """
    list1 = row['name'].split()
    for el in list1:
        try:
            year = int(el)
            if year >= 1980:
                return year - 1
        except ValueError: 
            continue
```

```python
# проверка работоспособности функции
df.query('year_of_release.isna()').apply(search_of_year, axis=1).head()
```

Заполним пропуски с помощью функции `search_of_year`

```python
#заполнение пропусков year_of_release
df['year_of_release'] = df['year_of_release'].fillna(df.apply(search_of_year, axis=1))
```

```python
#провека работы функции на отдельных значениях
df.loc[[183, 377, 475], :]
```

Далеко не для все данные удалось получить. Для остальных строк нет возможности восстановить год выпуска. (Можно попробовать гуглить в ручную, но камон) Заполним пропуски техническим значением 0 и сразу заменим на тип `int`.

```python
# заполнение пропусков в столбце year_of_release
df['year_of_release'] = df['year_of_release'].fillna(0)
```

```python
# замета типа данных на целочисленный
df['year_of_release'] = df['year_of_release'].astype('int')
```

## Проверка гипотез

### Пользовательские рейтинги платформ

Проверим следующую гипотезу "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые".

Для начала сделаем сводные таблицы для двух платформ

```python
xone = current_data.query('platform == "XOne" and user_score != -1 and not user_score.isna()')
pc = current_data.query('platform == "PC" and user_score != -1 and not user_score.isna()')
```

Это гипотеза о равенстве среднего двух генеральных совокупностей по взятым из них выборкам. Сформулируем  нулевую гипотезу в соответствии с правилом "Н0 всегда формулируется так, чтобы использовать знак равенства":

$H_0$: X1 = X2, т.е. средние пользовательские рейтинги платформ Xbox One и PC одинаковые

$H_1$: X1 $\neq$ X2, т.е. средние пользовательские рейтинги платформ Xbox One и PC разные

Таким образом, в данном случае  имеем дело с двусторонним статистическим критерием. Зададим уровень значимости 5%.

```python
# критический уровень статистической значимости
alpha = 0.05  
```

```python
results = st.ttest_ind(xone['user_score'], pc['user_score'], equal_var=False)
```

```python
results.pvalue 
```

```python
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

**Вывод**

Не удалось отвергнуть гипотезу о том, что  средние пользовательские рейтинги платформ Xbox One и PC одинаковые

### пользовательские рейтинги жанров 

```python
action = current_data.query('genre == "Action" and user_score != -1 and not user_score.isna()')
sports = current_data.query('genre == "Sports" and user_score != -1 and not user_score.isna()')
```

$H_0$: X1 = X2, т.е. средние пользовательские рейтинги жанров Action  и Sports одинаковые

$H_1$: X1 $\neq$ X2, т.е. средние пользовательские рейтинги жанров Action  и Sports разные

Таким образом, в данном случае  имеем дело с двусторонним статистическим критерием. Зададим уровень значимости 5%.
